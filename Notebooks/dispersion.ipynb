```python
import math

def number_of_Nash_equilibria(N, B):
    return math.comb(N, B)

def number_of_states(N):
    return 2 ** N
```

```python
import pandas as pd

max_N = 50
mus_values = [0.25, 0.5, 0.75]
Ns = list()
mus = list()
ne = list()
for i, mu in enumerate(mus_values):
    for N in range(2, max_N):
        Ns.append(N)
        B = int(mu * N)
        mus.append(mu)
        percentage_of_Nash_equilibria = number_of_Nash_equilibria(N, B) / number_of_states(N) * 100 
        ne.append(percentage_of_Nash_equilibria)

df = pd.DataFrame({
    'N': Ns,
    'mu': mus,
    '%NE': ne
})
df.head(2)
```

```python
import seaborn as sns

ax = sns.lineplot(data=df, x='N', y='%NE', hue='mu', style='mu', palette="tab10", linewidth=2.5)
ax.set_title(r"Percentage of Nash equilibria with N players and bar's capacity $\mu$")
ax.set_xlabel('Number of players N')
ax.set_ylabel('Percentage of Nash equilibria (%)')
ax.legend(title=r'$\mu$')
ax.grid()
```
