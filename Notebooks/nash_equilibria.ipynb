# Nash equilibria

This generates images of different Nash Equilibria in El Farol Bar Problem.

```python
import sys
sys.path.append("../src")

import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt

from Classes.bar import Bar
from Utils.cherrypick_simulations import CherryPickEquilibria
```

```python
eq = CherryPickEquilibria(
    num_agents=100,
    threshold=0.6,
    epsilon=0,
    num_rounds=100,
    num_episodes=1
)
eq.allow_shuffle = False
```

```python
df1 = eq.generate_segmentation_simulation()
bar = Bar.from_pandas(df1)

fig, ax = plt.subplots(figsize=(8, 8))

bar.render(ax=ax, num_rounds=len(bar.history))
plt.show()
```

```python
df2 = eq.generate_alternation_simulation()
bar = Bar.from_pandas(df2)

fig, ax = plt.subplots(figsize=(8, 8))

bar.render(ax=ax, num_rounds=len(bar.history))
plt.show()
```

```python
df3 = eq.generate_mixed_simulation()
bar = Bar.from_pandas(df3)

fig, ax = plt.subplots(figsize=(8, 8))

bar.render(ax=ax, num_rounds=len(bar.history))
plt.show()
```
