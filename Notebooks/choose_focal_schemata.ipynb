# Cherrypick focal regions for models

```python
import numpy as np

import sys

sys.path.append('../src')

from Config.config import PATHS
from Classes.focal_regions import SetFocalRegions, FocalRegion

import warnings
warnings.filterwarnings("ignore")
```

### Two players

**All regions**

```python
sfr = SetFocalRegions(
    num_agents=2,
    threshold=0.5,
    len_history=2,
    max_regions=10,
    from_file=False
)
sfr.generate_focal_regions()
for i, region in enumerate(sfr.focal_regions):
    FocalRegion.draw_region(
        region=region.focal_region,
        title=f"index {i}"
    )
```

**Cherrypicked regions**

```python
idxs = [0, 2]
sfr.focal_regions = list(np.array(sfr.focal_regions)[idxs])
for i, region in enumerate(sfr.focal_regions):
    FocalRegion.draw_region(
        region=region.focal_region,
        title=f"index {i}"
    )
```

**Save regions**

```python
sfr.save_focal_regions()
```

**Test**

```python
sfr.focal_regions = sfr.load_focal_regions()
for i, region in enumerate(sfr.focal_regions):
    FocalRegion.draw_region(
        region=region.focal_region,
        title=f"index {i}"
    )
```
